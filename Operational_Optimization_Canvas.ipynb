```python
import numpy as np, matplotlib.pyplot as plt

stages = ["Input","Processing","Review","Delivery"]
baseline = np.random.randint(60,100,len(stages))
optimized = baseline - np.random.randint(10,30,len(stages))
plt.plot(stages, baseline, 'r--o', label="Before AI")
plt.plot(stages, optimized, 'g-o', label="After AI")
plt.title("Operational Optimization Canvas â€“ Workflow Efficiency")
plt.ylabel("Task Time (%)"); plt.legend(); plt.show()
```
