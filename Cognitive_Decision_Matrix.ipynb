```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt

criteria = ["Impact","Feasibility","Cost","Risk"]
scores = np.random.randint(60,95,len(criteria))
weights = [0.4,0.3,0.2,0.1]
decision_score = np.dot(scores,weights)
plt.bar(criteria,scores)
plt.title(f"Cognitive Decision Matrix â€“ Score = {decision_score:.1f}")
plt.ylabel("Criterion Score"); plt.show()
```
